```python
print('**************')

region = int(input('''To which area would you like to make a call? (additional fees may apply)
1 - Asia          (+5)
2 - Europe        (+15)
3 - Australia     (+5)
4 - North America (+10)
5 - South America (+10)
6 - Africa        (+15)
7 - Local         ----
                   \nPlease enter the corresponding digit:
'''))
print('--------------')
time = int(input('''\nTime of call (additional fees may apply):
1 - Day
2 - Night (+5)
\nPlease enter the corresponding digit:
'''))
print('--------------')
duration = int(input('''\nEnter duration of call (in minutes): 
After a minute, an additional fee of Php 5 is added per succeeding minute.
'''))
#service fee - 10
fee = 10
if region == 1 or region == 3:
    fee += 5
elif region == 4 or region == 5:
    fee += 10
elif region == 2 or region == 6:
    fee += 15

if time == 2:
    fee += 5

if duration > 1:
    fee += ((duration - 1)*5) 
print(f'The total fee (service + additional fees) for the call is Php {fee}')
print(f'\nThank you for using our service! We hope you are satisfied.')
print('**************')
```
